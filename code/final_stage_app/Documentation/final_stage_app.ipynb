## Overview of GUI Application

The root folder of my repository for the pands project has a 'code' folder.  
Within this code folder is the "sub-project" called **final_stage_app**  
  
![alt text](../../../images/pands_github.png)  
  
This folder has several other subfolders ...
* data
* images
* logs
* notebooks



Located in the root of this **final_stage_app** folder are the  original file generated is called [main.py](../main.py) and supporting modules. They were all converted into an OOP based class.  
https://realpython.com/python3-object-oriented-programming/  
https://www.geeksforgeeks.org/python-oops-concepts/  
There a several other Classes that act as supporting modules for this application.  

* **datasets.py**  
This is a utility for loading and Reading csv-based datasets   
* **ploys.py**  
This is a utility for plotting Matplotlib based graphs and outputting generated images to a jupyter notebook.  
* **app_config.py**  
This is a utility for managing the configuration of application. It is a simplified wrapper for the python configparser module  
https://docs.python.org/3/library/configparser.html  
https://www.tutorialspoint.com/configuration-file-parser-in-python-configparser  
* **app_logger.py**   
This is a utility for managing the logging of of application events. It is a simplified wrapper for the python configparser module  
https://docs.python.org/3/library/logging.html
https://realpython.com/python-logging/

1. Initially, when application launches the data folder is parsed for *.csv files.  
1. The files within this folder populate the top combobox control of form.
1. When the application user selects which dataset to process and clicks 'Load Data', the lower listbox is populated with the attributes features of that dataset.
1. The user then selects one or all of ..
    * a text-based Statistical Summary
    * a Histogram
    * a Boxplot
1. finally, the user clicks 'Generate'

A Jupyter Note book named Analysis_XXXXXX.ipynb is generated with in 'notebooks' folder
