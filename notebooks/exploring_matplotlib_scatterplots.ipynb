# Exploring Matplotlib's Scatter Plots



In this notebook, we explore the application of the Scatter Plot to visualize the relationship between 2 variables

According to Michael Friendly and Daniel Denis in their 2005 paper [The early origins and development of the scatterplot](https://pubmed.ncbi.nlm.nih.gov/15812820/), the first use of a scatter plot is attributed to John Herschel In 1833 who plotted the movements of astronomical Stars over time.   

> "Of all the graphic forms used today, the scatterplot is arguably the most versatile, polymorphic, and generally useful invention in the history of statistical graphics." 
  
*With Scatter Plots Herschel could use freehand drawing and human judgment to extract the information he required.  
Analyzing Scatter Plots can be an exercise in Subjective visual assessment of data rather than objective calculation.*

# Using Matplotlib 
### Start with a look at plotting a basic scatter plot of _Random Points_ for now...   

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)  # Fixing random state for reproducibility
x, y = np.random.randn(2, 400) * [[.5], [180]]

plt.scatter(x, y)

plt.show()
```

## Lets try a 3D Scatter plot ...

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)  # Fixing random state for reproducibility
x, y, z = np.random.randn(3, 400) * [[.5], [180], [20]]


fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x, y, z)
#remove axis ticks
ax.set(xticklabels=[], yticklabels=[], zticklabels=[])
#lets not bother with axis labels, ticks, titles for now

plt.show()
```

## Add Random Color and Marker Size ...

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)  # Fixing random state for reproducibility
x, y, z = np.random.randn(3, 400) * [[.5], [.5], [.5]]

colors = np.random.randint(100, size=(400))
sizes = 1 * np.random.randint(200, size=(400))

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x, y, z, s=sizes, c=colors)
#remove axis ticks
ax.set(xticklabels=[], yticklabels=[], zticklabels=[])
#lets not bother with axis labels, ticks, titles for now

plt.show()
```

## Try Impose a 2D Histogram on a 2D Scatter plot

Encountered an interesting plot while researching scatter plots ... both a scatter plot and histogram 
![alt text](../images/scatter_histo.png)

As per previous [Notebook](exploring_matplotlib_layouts.ipynb), let use Grid Spec.   
(Used ChatGPT to guide me thru code. Use of gridspec was a little confusing ... *Remember it is Rows, Columns and not X,Y*)

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

np.random.seed(19680801)  # Fixing random state for reproducibility
x, y = np.random.randn(2, 400) * [[.5], [.5]]
colors = np.random.randint(100, size=(400))
sizes = 1 * np.random.randint(200, size=(400))

# Set up the layout 
fig = plt.figure(figsize=(8, 8))
gs = gridspec.GridSpec(4, 4)

#Setup Axes that will merge across the 4x4 grid
ax_scatter = plt.subplot(gs[1:4, 0:3])  # Plot spans x grid 1-4 and y grid 0-3 for a 3x3 spanned plot
ax_xhist = plt.subplot(gs[0, 0:3], sharex=ax_scatter)  # Histogram spans Columns 0-3 at row #0
ax_yhist = plt.subplot(gs[1:4, 3], sharey=ax_scatter)  # Histogram spans Rows 1-4 at Column #3

# Scatter plot
ax_scatter.scatter(x, y, alpha=0.6, c=colors, s=sizes)

# Histograms
ax_xhist.hist(x, bins=20, color='skyblue',edgecolor = 'white')
ax_yhist.hist(y, bins=20, orientation='horizontal', color='skyblue',edgecolor = 'white')

# Hide tick labels on histograms
ax_xhist.set_xticklabels([])
ax_xhist.set_yticklabels([])
ax_xhist.set_yticklabels([])
ax_yhist.set_yticklabels([])
# Adjust spacing
plt.tight_layout()
```

### Use of Seaborn for simplier code

A quick look at the use of Seaborn modules to generate Scatter plots and histograms.  
*Time constraints prevented me from expanding plot to use colored, or variable sized, Markers*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(19680801)  # Fixing random state for reproducibility
x, y = np.random.randn(2, 400) * [[.5], [.5]]


# Create the joint plot
ax = sns.jointplot(x=x, y=y, kind='scatter', marginal_kws=dict(bins=20, fill=True))
ax.ax_joint.set_xlabel("X Values")
ax.ax_joint.set_ylabel("Y Values")

plt.show()
```
